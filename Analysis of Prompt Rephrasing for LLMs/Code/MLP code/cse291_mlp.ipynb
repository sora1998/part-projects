```python
import re
def preprocess(df):
    data={}
    ci=['A','B','C','D','E','F','G']
    data['question']=list(cp.deepcopy(df['question']))
    data['choices']=list(cp.deepcopy(df['choices']))
    data['answer']=list(cp.deepcopy(df['answer']))
    data['type']=list(cp.deepcopy(df['type']))
    data['choices']=list(cp.deepcopy(df['choices']))
    for i in range(len(data['answer'])):
        temp=""
        choices=re.split("\",|\',",data['choices'][i].replace("[","").replace("]",""))
        for j in range(len(choices)):
            choices[j]=choices[j].replace("'","").replace('"','')
        data['answer'][i]=ci[int(data['answer'][i])]
        for k in range(len(choices)):
            if k==0:
                temp+=" [ "
            temp=temp+" "+ci[k]+": \""+choices[k]+"\" "
            if k==len(choices)-1:
                temp+=" ] "
        data['choices'][i]=temp
    return data
```

```python
import re
def process_answer(ans):
    ans=ans.replace("'","").replace('"',"").replace("{","").replace("}","")
    if "Question" in ans:
        ans=ans.replace("Question","").strip()
    if "\n" in ans:
        temp=ans.split("\n")
    elif "," in ans:
        temp=ans.split(",")
    elif '\",\"' in ans:
        temp=ans.split('\",\"')
    elif '\", \"' in ans:
        temp=ans.split('\", \"')
    elif ";" in ans:
        temp=ans.split(";")
    for i in range(len(temp)):
        if temp[i]=="":
            continue
        elif (":" in temp[i]) and re.sub(r'[^\w\s]','',(temp[i].split(":")[0])).strip().isnumeric():
            try:
                temp[i]=(temp[i].split(":")[0],temp[i].replace(temp[i].split(":")[0]+":","").strip())
            except:
                continue
        elif (":\"" in temp[i]) and re.sub(r'[^\w\s]', '',(temp[i].split(":")[0])).strip().isnumeric():
            try:
                temp[i]=(temp[i].split(":")[0],temp[i].replace(temp[i].split(":")[0]+":","").strip())
            except:
                continue
        elif ('-' in temp[i]) and re.sub(r'[^\w\s]', '',(temp[i].split("-")[0])).strip().isnumeric():
            try:
                temp[i]=(temp[i].split("-")[0],temp[i].replace(temp[i].split("-")[0]+"-","").strip())
            except:
                continue
        elif ('-' in temp[i]) and re.sub(r'[^\w\s]', '',(temp[i].split("-")[0])).strip().isnumeric():
            try:
                temp[i]=(temp[i].split("-")[0],temp[i].replace(temp[i].split("-")[0].strip()+" -","").strip())
            except:
                continue
        elif (')' in temp[i]) and re.sub(r'[^\w\s]', '',(temp[i].split(")")[0])).strip().isnumeric():
            try:
                temp[i]=(temp[i].split(")")[0],temp[i].replace(temp[i].split(")")[0]+")","").strip())
            except:
                continue
        else:
            try:
                if temp[i].split(".")[0].strip().isnumeric():
                    temp[i]=(temp[i].split(".")[0],temp[i].replace(temp[i].split(".")[0]+".",""))
                    temp0=temp[i][0]
                    temp1=temp[i][1].split("-")[0].strip()
                    temp[i]=(temp0,temp1)
                else:
                    temp[i]=(temp[i].split(".")[0].strip(),temp[i].split(".")[0].strip())
            except:
                continue
        if "Option" in temp[i][1]:
            temp0=temp[i][0]
            temp1=temp[i][1].split(":")[0].replace("Option","").strip()
            temp[i]=(temp0,temp1)
        elif "Answer" in temp[i][1]:
            temp0=temp[i][0]
            temp1=temp[i][1].split(":")[1].strip()[0]
            temp[i]=(temp0,temp1)
        elif ":" in temp[i][1]:
            temp0=temp[i][0]
            temp1=temp[i][1].split(":")[0].strip()
            temp[i]=(temp0,temp1)
        elif "-" in temp[i][1]:
            temp0=temp[i][0]
            temp1=temp[i][1].split("-")[0].strip()
            temp[i]=(temp0,temp1)
    temp=[data for data in temp if data!=""]
    return temp
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from functools import reduce
from textblob import TextBlob
```

```python
def check_answer(result,data):
    total=0
    correct=0
    wrong_question=[]
    ps = PorterStemmer()
    for i in range(len(result)):
        total+=1
        if len(result[i])>2:
            s1=nltk.word_tokenize(result[i][1] .lower())
            s2=nltk.word_tokenize(data['answer'][i].lower())
            s1=[k for k in s1 if k not in ["'s","'",'"',"``","''","\\","\\'s"]]
            s2=[k for k in s2 if k not in ["'s","'",'"',"``","''","\\","\\'s"]]
            s1=reduce(lambda x, y: x + " " + ps.stem(y), s1, "")
            s2=reduce(lambda x, y: x + " " + ps.stem(y), s2, "")
            s1=TextBlob(s1).correct()
            s2=TextBlob(s2).correct()
            s1=re.sub(r'[^\w\s]','',str(s1))
            s2=re.sub(r'[^\w\s]','',str(s2))
            c_correct=0
            for j in range(len(s2)):
                if j <len(s1):
                    if s1[j]==s2[j]:
                        c_correct+=1
            if c_correct/len(s2) > 0.95:
                correct+=1
            else:
                wrong_question.append((data['question'][i],data['choices'][i],data["answer"][i],result[i],data['type'][i]))
            continue
        else:
#             s1 = ''.join(c for c in result[i][1] .lower() if c!=" ")
#             s2 = ''.join(c for c in data['answer'][i].lower() if c!=" ")
            s1=nltk.word_tokenize(result[i][1] .lower())
            s2=nltk.word_tokenize(data['answer'][i].lower())
            s1=[k for k in s1 if k not in ["'s","'",'"',"``","''","\\","\\'s"]]
            s2=[k for k in s2 if k not in ["'s","'",'"',"``","''","\\","\\'s"]]
            s1=reduce(lambda x, y: x + " " + ps.stem(y), s1, "")
            s2=reduce(lambda x, y: x + " " + ps.stem(y), s2, "")
            s1=TextBlob(s1).correct()
            s2=TextBlob(s2).correct()
            s1=re.sub(r'[^\w\s]','',str(s1))
            s2=re.sub(r'[^\w\s]','',str(s2))
            c_correct=0
            for j in range(len(s2)):
                if j <len(s1):
                    if s1[j]==s2[j]:
                        c_correct+=1
            if c_correct/len(s2) > 0.95:
                correct+=1
            else:
                wrong_question.append((data['question'][i],data['choices'][i],data["answer"][i],result[i][1],data['type'][i]))
    accuracy=correct/len(result)
    return correct,accuracy,wrong_question
```

```python
def check_accuracy_each_type(chat_dict,true_dict):
    result=defaultdict(float)
    for key in true_dict.keys():
        if key in chat_dict.keys():
            result[key]=(true_dict[key]-chat_dict[key])/true_dict[key]
        else:
            result[key]=1
    return result
```
